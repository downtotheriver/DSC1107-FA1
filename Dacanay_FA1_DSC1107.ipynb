# **Variation**

## **Discrete Variables**

```r
species_count <- penguins %>%
    count(species) %>%
    arrange(n) 

ggplot(species_count, aes(x = reorder(species, n), y = n, fill = species)) +
    geom_bar(stat = "identity") +
    labs(x = "Species", y = "Count", title = "Distribution of Penguin Species") +
    scale_fill_manual(values = c("Adelie" = "#377EB8", "Chinstrap" = "#4DAF4A", "Gentoo" = "#FF7F00")) +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

## **Continuous Variables**

```r
bin = (max(penguins$body_mass_g, na.rm=TRUE) - min(penguins$body_mass_g, na.rm=TRUE)) / 15

ggplot(data = penguins) +
    geom_histogram(mapping = aes(x = body_mass_g), binwidth = bin) +
    labs(title = "Histogram of Penguins' Body Mass", x = "Body Mass", y = "Count") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

```r
ggplot(data = penguins) +
    geom_histogram(mapping = aes(x = body_mass_g), binwidth = bin) +
    labs( x = "Body Mass", y = "Count") +
    coord_cartesian(xlim = c(2850,6040))
```

# **Covariation**

## **Discrete vs. Discrete**

```r
ggplot(data = penguins) +
    geom_count(mapping = aes(x = species, y = island)) +
    ggtitle("Count of Penguins by Species and Island") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

## **Continuous vs. Continuous**

### **Many to one**

```r
ggplot(data = penguins) +
    geom_point(mapping = aes(x = bill_length_mm, y = bill_depth_mm)) +
    ggtitle("Measurement of Penguins' Bill Depth
per Bill Length (mm)") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

### **One to one**

```r
avg_flipper_length_by_body_mass <- penguins %>%
    mutate(body_mass = floor(body_mass_g/100) * 100) %>%
    group_by(body_mass) %>%
    summarise(avg_flipper_length = mean(flipper_length_mm, na.rm=TRUE))
avg_flipper_length_by_body_mass
```

```r
ggplot(data = avg_flipper_length_by_body_mass) +
    geom_line(mapping = aes(x = body_mass, y = avg_flipper_length)) +
    ggtitle("Penguin's Average Flipper Length
per Group of Body Mass") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

## **Continuous vs. Discrete**

```r
ggplot(data = penguins) +
    geom_boxplot(mapping = aes(x = species, y = body_mass_g)) +
    ggtitle("Boxplot of Penguins' Body Mass per Species") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

## **Visualization**
### **Axis transformations**

```r
ggplot(data = penguins) +
    geom_histogram(aes(x = body_mass_g), binwidth = bin) +
    ggtitle("Histogram of Penguins' Body Mass") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

```r
ggplot(data = penguins) +
    geom_histogram(aes(x = body_mass_g), binwidth = bin) +
    scale_x_log10() +
    ggtitle("Histogram of Penguins' Body Mass
with X-axis on a Logarithmic Scale") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

```r
ggplot(data = penguins) +
    geom_point(aes(x = body_mass_g, y = flipper_length_mm)) +
    scale_x_log10() +
    scale_y_log10() +
    ggtitle("Penguins' Body Mass vs. Flipper Length") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

### **Multiple geoms in the same plot**

```r
ggplot(data = avg_flipper_length_by_body_mass) +
    geom_line(mapping = aes(x = body_mass, y = avg_flipper_length)) +
    geom_point(mapping = aes(x = body_mass, y = avg_flipper_length)) +
    ggtitle("Penguin's Average Flipper Length
per Group of Body Mass") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

```r
ggplot(data = avg_flipper_length_by_body_mass, mapping = aes(x = body_mass, y = avg_flipper_length)) +
    geom_line() +
    geom_point()
```

### **Multiple aesthetics in the same geom**

```r
ggplot(data = penguins) +
    geom_point(mapping = aes(x = body_mass_g, y = flipper_length_mm, color = species)) +
    ggtitle("Scatter Plot of Body Mass and Flipper Length
by Penguin Species") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

### **Faceting**

```r
ggplot(data = penguins) +
    geom_point(mapping = aes(x = body_mass_g, y = flipper_length_mm)) +
    facet_wrap(~species) +
    ggtitle("Scatter Plot of Body Mass and Flipper Length
by Penguin Species") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

```r
ggplot(data = penguins) +
    geom_point(mapping = aes(x = body_mass_g, y = flipper_length_mm)) +
    facet_grid(island ~ species) +
    ggtitle("Scatter Plot of Body Mass and Flipper Length
by Penguin Species and Island") +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```
